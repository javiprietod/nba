```python
import requests
```

```python
# datos cogidos de https://sportsdata.io/developers/api-documentation/nba
response = requests.get('https://api.sportsdata.io/v3/nba/scores/json/Games/2022?key=5ecd3d649b5f4b51a6cb172b11cba307').json()

```

```python
len(response)
teams = {}
for game in response:
    if game['AwayTeam'] not in teams:
        teams[game['AwayTeam']] = game['AwayTeamID']
teams
```

```python
# wins in the season 2021-2022
wins = 0
for i in response:
    if i['AwayTeam'] == 'GS':
        if i['AwayTeamScore'] > i['HomeTeamScore']:
            wins += 1
    elif i['HomeTeam'] == 'GS':
        if i['HomeTeamScore'] > i['AwayTeamScore']:
            wins += 1
wins
```
